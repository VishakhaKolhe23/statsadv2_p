**Statistics Part 2 - Theory Questions**

1. **What is hypothesis testing in statistics?**
   Hypothesis testing is a statistical method that allows us to make inferences or draw conclusions about a population based on sample data. It involves testing an assumption (hypothesis) regarding a population parameter.

2. **What is the null hypothesis, and how does it differ from the alternative hypothesis?**
   The **null hypothesis (H₀)** assumes no effect or no difference, while the **alternative hypothesis (H₁ or Ha)** assumes there is an effect or a difference.

3. **What is the significance level in hypothesis testing, and why is it important?**
   The **significance level (α)** is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 or 0.01. It determines the threshold for statistical significance.

4. **What does a P-value represent in hypothesis testing?**
   A **P-value** measures the probability of obtaining results as extreme as the observed results, under the assumption that the null hypothesis is true.

5. **How do you interpret the P-value in hypothesis testing?**
   If **P-value < α**, reject H₀ (evidence supports H₁). If **P-value ≥ α**, fail to reject H₀ (insufficient evidence for H₁).

6. **What are Type 1 and Type 2 errors in hypothesis testing?**

   * **Type 1 Error**: Rejecting H₀ when it is true (false positive).
   * **Type 2 Error**: Failing to reject H₀ when it is false (false negative).

7. **What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**

   * **One-tailed test**: Tests for effect in one direction.
   * **Two-tailed test**: Tests for effect in both directions.

8. **What is the Z-test, and when is it used in hypothesis testing?**
   Z-test is used when the population variance is known and the sample size is large (n > 30).

9. **How do you calculate the Z-score, and what does it represent in hypothesis testing?**
   $Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}$
   Represents how many standard deviations the sample mean is from the population mean.

10. **What is the T-distribution, and when should it be used instead of the normal distribution?**
    The T-distribution is used when the population standard deviation is unknown and the sample size is small (n < 30).

11. **What is the difference between a Z-test and a T-test?**

    * **Z-test**: Known population variance.
    * **T-test**: Unknown population variance.

12. **What is the T-test, and how is it used in hypothesis testing?**
    The T-test compares sample means when population variance is unknown. Types: one-sample, independent two-sample, and paired sample T-tests.

13. **What is the relationship between Z-test and T-test in hypothesis testing?**
    Both are used to test mean differences, but the T-test is more flexible when variance is unknown.

14. **What is a confidence interval, and how is it used to interpret statistical results?**
    A confidence interval gives a range within which we expect the true population parameter to fall with a certain confidence level (e.g., 95%).

15. **What is the margin of error, and how does it affect the confidence interval?**
    The margin of error is the range added and subtracted from the sample mean to create the confidence interval. Larger margin = wider interval.

16. **How is Bayes' Theorem used in statistics, and what is its significance?**
    Bayes' Theorem updates the probability of a hypothesis based on new evidence. It is essential for Bayesian inference.

17. **What is the Chi-square distribution, and when is it used?**
    A distribution used for categorical data analysis, particularly in tests of independence and goodness-of-fit.

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)

# 1. Independent T-test
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
t_stat, p_val = stats.ttest_ind(group1, group2)
print(f"Independent T-test: t-statistic = {t_stat:.3f}, p-value = {p_val:.3f}")
print("Interpretation:", "Reject H0" if p_val < 0.05 else "Fail to reject H0")

# 2. Paired sample T-test and visualization
before = np.random.normal(100, 15, 25)
after = before + np.random.normal(5, 10, 25)
t_stat_paired, p_val_paired = stats.ttest_rel(before, after)
print(f"\nPaired T-test: t-statistic = {t_stat_paired:.3f}, p-value = {p_val_paired:.3f}")

plt.figure(figsize=(8,4))
plt.plot(before, label='Before', marker='o')
plt.plot(after, label='After', marker='o')
plt.title("Paired Sample Comparison")
plt.xlabel("Sample Index")
plt.ylabel("Values")
plt.legend()
plt.show()

# 3. Simulate data and perform Z-test and T-test
# For Z-test, population std known, use large sample
pop_mean = 100
pop_std = 15
sample = np.random.normal(pop_mean, pop_std, 100)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

# Z-test statistic
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_val_z = 2 * (1 - stats.norm.cdf(abs(z)))
print(f"\nZ-test: z = {z:.3f}, p-value = {p_val_z:.3f}")

# T-test statistic (pop std unknown)
t_stat_sample, p_val_sample = stats.ttest_1samp(sample, pop_mean)
print(f"T-test: t = {t_stat_sample:.3f}, p-value = {p_val_sample:.3f}")

# 4. Confidence interval function for sample mean
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)
    h = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)
    return mean - h, mean + h

ci_lower, ci_upper = confidence_interval(sample)
print(f"\n95% Confidence Interval for sample mean: ({ci_lower:.2f}, {ci_upper:.2f})")
print("Significance: We are 95% confident that the true population mean lies within this interval.")

# 5. Margin of error calculation
def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_err = stats.sem(data)
    margin = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)
    return margin

moe = margin_of_error(sample)
print(f"\nMargin of Error (95%): ±{moe:.2f}")

# 6. Bayesian inference example using Bayes' Theorem
# P(A|B) = P(B|A)*P(A)/P(B)
def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    p_not_a = 1 - p_a
    p_b = p_b_given_a * p_a + p_b_given_not_a * p_not_a
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b

# Example: Disease prevalence and test accuracy
p_disease = 0.01
p_pos_given_disease = 0.99
p_pos_given_no_disease = 0.05
posterior = bayes_theorem(p_disease, p_pos_given_disease, p_pos_given_no_disease)
print(f"\nBayesian Inference: Probability of disease given positive test = {posterior:.3f}")

# 7. Chi-square test of independence
# Create contingency table
data = pd.DataFrame({
    'Smoker': ['Yes', 'Yes', 'No', 'No'],
    'Disease': ['Yes', 'No', 'Yes', 'No'],
    'Count': [30, 70, 20, 80]
})
contingency = pd.pivot_table(data, values='Count', index='Smoker', columns='Disease')
chi2, p, dof, expected = stats.chi2_contingency(contingency)
print(f"\nChi-square test of independence: chi2 = {chi2:.3f}, p = {p:.3f}")

# 8. Calculate expected frequencies (already done above as 'expected')
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=contingency.index, columns=contingency.columns))

# 9. Goodness-of-fit test
observed = np.array([30, 10, 25, 35])
expected_freq = np.array([25, 25, 25, 25])
chi2_gof, p_gof = stats.chisquare(f_obs=observed, f_exp=expected_freq)
print(f"\nGoodness-of-fit test: chi2 = {chi2_gof:.3f}, p = {p_gof:.3f}")

# 10. Simulate and visualize Chi-square distribution
df_chi = 4
x = np.linspace(0, 15, 100)
y = stats.chi2.pdf(x, df_chi)
plt.figure()
plt.plot(x, y, label=f'Chi-square df={df_chi}')
plt.title("Chi-square Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()

# 11. F-test to compare variances
sample1 = np.random.normal(50, 10, 30)
sample2 = np.random.normal(50, 15, 30)
f_stat = np.var(sample1, ddof=1) / np.var(sample2, ddof=1)
df1 = len(sample1) - 1
df2 = len(sample2) - 1
p_f = 1 - stats.f.cdf(f_stat, df1, df2)
print(f"\nF-test: F-statistic = {f_stat:.3f}, p-value = {p_f:.3f}")

# 12. ANOVA test and interpretation
group_a = np.random.normal(60, 8, 25)
group_b = np.random.normal(65, 8, 25)
group_c = np.random.normal(70, 8, 25)
f_stat_anova, p_anova = stats.f_oneway(group_a, group_b, group_c)
print(f"\nANOVA test: F = {f_stat_anova:.3f}, p = {p_anova:.3f}")
print("Interpretation:", "Significant difference between groups" if p_anova < 0.05 else "No significant difference")

# 13. One-way ANOVA with plot
df_anova = pd.DataFrame({
    'value': np.concatenate([group_a, group_b, group_c]),
    'group': ['A']*25 + ['B']*25 + ['C']*25
})
plt.figure()
sns.boxplot(x='group', y='value', data=df_anova)
plt.title("One-way ANOVA Groups Comparison")
plt.show()

# 14. Check ANOVA assumptions function
def check_anova_assumptions(data, group_col, value_col):
    groups = [data[data[group_col]==grp][value_col] for grp in data[group_col].unique()]
    # Normality (Shapiro-Wilk)
    normality = [stats.shapiro(g)[1] > 0.05 for g in groups]
    # Equal variance (Levene's test)
    equal_var = stats.levene(*groups)[1] > 0.05
    return normality, equal_var

normality, equal_var = check_anova_assumptions(df_anova, 'group', 'value')
print(f"\nANOVA Assumptions:\nNormality per group: {normality}\nEqual Variance: {equal_var}")

# 15. Two-way ANOVA (simplified example using statsmodels)
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Simulate data with two factors
np.random.seed(0)
factor1 = np.repeat(['Low', 'High'], 30)
factor2 = np.tile(np.repeat(['Type1', 'Type2'], 15), 2)
value = 10 + (factor1 == 'High')*2 + (factor2 == 'Type2')*3 + np.random.normal(0, 1, 60)

df_two_way = pd.DataFrame({'Factor1': factor1, 'Factor2': factor2, 'Value': value})
model = ols('Value ~ C(Factor1) * C(Factor2)', data=df_two_way).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("\nTwo-way ANOVA results:\n", anova_table)

# Interaction plot
from statsmodels.graphics.factorplots import interaction_plot
plt.figure()
interaction_plot(df_two_way['Factor1'], df_two_way['Factor2'], df_two_way['Value'])
plt.title("Interaction Plot for Two-way ANOVA")
plt.show()

# 16. Visualize F-distribution
x_f = np.linspace(0, 5, 100)
y_f = stats.f.pdf(x_f, dfn=5, dfd=2)
plt.figure()
plt.plot(x_f, y_f)
plt.title("F-distribution (dfn=5, dfd=2)")
plt.xlabel("F value")
plt.ylabel("Density")
plt.show()

# 17. One-way ANOVA boxplot already covered in #13

# 18. Simulate normal data & hypothesis test (one-sample)
data_norm = np.random.normal(50, 5, 40)
t_stat_norm, p_val_norm = stats.ttest_1samp(data_norm, 50)
print(f"\nHypothesis test on normal data: t = {t_stat_norm:.3f}, p = {p_val_norm:.3f}")

# 19. Chi-square test for population variance (using Chi-square distribution)
sample_var = np.var(data_norm, ddof=1)
n_var = len(data_norm)
alpha = 0.05
chi2_lower = stats.chi2.ppf(alpha/2, n_var-1)
chi2_upper = stats.chi2.ppf(1 - alpha/2, n_var-1)
ci_lower_var = (n_var-1)*sample_var/chi2_upper
ci_upper_var = (n_var-1)*sample_var/chi2_lower
print(f"\n95% CI for population variance: ({ci_lower_var:.2f}, {ci_upper_var:.2f})")

# 20. Z-test for comparing proportions
count1 = 40
nobs1 = 100
count2 = 30
nobs2 = 100
stat, pval = stats.proportions_ztest([count1, count2], [nobs1, nobs2])
print(f"\nZ-test for proportions: z = {stat:.3f}, p = {pval:.3f}")

# 21. F-test for variance comparison with visualization (already calculated in #11)
plt.figure()
sns.histplot(sample1, color='blue', label='Sample 1', kde=True)
sns.histplot(sample2, color='red', label='Sample 2', kde=True)
plt.title('Distribution Comparison for F-test')
plt.legend()
plt.show()

# 22. Chi-square goodness-of-fit with simulated data (already done in #9)

```
